## 4. Summarizing Data

In this section, we aim to explore the relationship between game genre and sales performance across various platforms using the video_games.csv dataset. Our research question focuses on how different video game genres influence the sales performance of games in the North American market. To address this, we will analyze the Genre (categorical variable) and US Sales (millions) (quantitative variable) to uncover any patterns or correlations between these variables. Visualizations and summary statistics will be used to highlight key trends and to provide a comprehensive understanding of the sales dynamics for different genres.

```python
sales_by_genre = df_critical.groupby('Genre')['US Sales (millions)'].sum().reset_index()

plt.figure(figsize=(15,15))
sns.barplot(data=sales_by_genre, x='US Sales (millions)', y='Genre', estimator=sum)
plt.title('Total Global Sales by Game Genre')
plt.xticks(rotation=90)
plt.show()
```

The bar plot reveals that the “Action” genre, especially when combined with sub-genres like “Adventure” or “Racing/Driving,” overwhelmingly dominates global video game sales, significantly surpassing other genres. This indicates the immense popularity and commercial success of action-based games. In contrast, genres like “Strategy” and “Simulation” exhibit much lower sales figures, suggesting they cater to more niche markets. Additionally, many games are categorized under multiple genre combinations, reflecting the diversity and complexity of game offerings.

```python
avg_sales_by_genre = df_critical.groupby('Genre')['US Sales (millions)'].median().reset_index()

plt.figure(figsize=(15, 15))
sns.barplot(data=avg_sales_by_genre, x='US Sales (millions)', y='Genre', palette='viridis')
plt.title('Average Sales per Game by Genre')
plt.xlabel('Average US Sales (millions)')
plt.ylabel('Genre')
plt.show()
```

```python
avg_sales_by_genre = df_critical.groupby('Genre')['US Sales (millions)'].mean().reset_index()

plt.figure(figsize=(15, 15))
sns.barplot(data=avg_sales_by_genre, x='US Sales (millions)', y='Genre', palette='viridis')
plt.title('Average Sales per Game by Genre')
plt.xlabel('Average US Sales (millions)')
plt.ylabel('Genre')
plt.show()
```

In the “Average Sales per Game by Genre” chart, the genre combination of Educational/Sports stands out with a significantly higher average sales figure compared to other genres. This suggests that games in this specific genre category are likely contributing disproportionately to overall sales, potentially due to the success of a few blockbuster titles that have driven up the average. This dominance could also indicate a specialied market where fewer games are produced, but the ones that are released capture a substantial portion of the market. The outsized performance of Educational/Sports games highlights how niche genres can sometimes outperform more common genres on a per-game basis, likely due to a combination of targeted audiences and high-quality, high-demand releases within this category.

```python
plt.figure(figsize=(15,15))
sns.boxplot(data=df_critical, x='US Sales (millions)', y='Genre')
plt.title('Sales Distribution by Genre')
plt.xticks(rotation=90)
plt.show()
```

The box plot provides insights into the distribution of global sales across different game genres. One of the key observations is that genres such as “Action” and “Shooter” exhibit a wide range of sales, with several outliers indicating that certain games in these genres achieve exceptionally high sales compared to others. This suggests variability within these genres, where a few blockbuster titles drive significant sales. On the other hand, genres like “Puzzle” and “Strategy” show much tighter sales distributions, reflecting more consistent, albeit lower, sales figures across the board. Additionally, some genres, such as “Sports,” show a relatively balanced distribution with fewer extreme outliers.

```python
sales_by_genre_platform = df_critical.groupby(['Genre', 'Console'])['US Sales (millions)'].sum().reset_index()

plt.figure(figsize=(15,50))
sns.barplot(data=sales_by_genre_platform, x='US Sales (millions)', y='Genre', hue='Console')
plt.title('Total Sales by Genre and Platform')
plt.xticks(rotation=90)
plt.show()
```

The bar plot illustrates the total sales of different game genres across various platforms, highlighting that “Action” and “Action-Adventure” dominate sales, particularly on the Xbox 360 and PlayStation 3. This shows that action-based games have wide appeal and perform strongly across multiple platforms. Certain genres, such as “Sports,” also show platform-specific strengths, with Xbox 360 leading in sports game sales. In contrast, handheld platforms like the Sony PSP and Nintendo DS show significantly lower sales across most genres, suggesting that home consoles outperform handhelds in many popular genres.
